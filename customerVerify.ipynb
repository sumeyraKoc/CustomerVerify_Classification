### Describe and implament a few helper functions for data preprocessing

```python
def percent_null_values_all_columns(df: pd.DataFrame) -> pd.DataFrame:
    """Calculate the percentage of null values in each column of a DataFrame.

    Args:
    - df (pd.DataFrame): The DataFrame to analyze.

    Returns:
    - pd.DataFrame: A DataFrame with two columns, 'Column' and 'Percent_Null', representing the percentage of null values in each column.
    """
    total_rows = len(df)
    null_percentage_data = []

    for column in df.columns:
        null_count = df[column].isnull().sum()
        null_percentage = (null_count / total_rows) * 100
        formatted_null_percentage = round(null_percentage, 2)
        null_percentage_data.append({'Feature': column, 'Percent_Null': formatted_null_percentage})


    null_percentage_df = pd.DataFrame(null_percentage_data)
    print(null_percentage_df)
```

```python
def unique_value_percentages_of_total(df: pd.DataFrame) -> pd.DataFrame:
    """Calculate the percentage of unique values relative to the sum of all observations for each column in a DataFrame.

    Args:
    - df(pd.DataFrame): The DataFrame to analyze.

    Returns:
    - pd.DataFrame: A DataFrame with columns and their corresponding percentage of unique values of total observations.
    """
    total_observations = len(df)
    percentages = []

    for column in df.columns:
        num_unique_values = df[column].nunique()
        percentage = (num_unique_values / total_observations) * 100
        formatted_percentage = round(percentage, 2)
        percentages.append({'Column': column, 'Percentage_of_Total':formatted_percentage})

    percentage_df = pd.DataFrame(percentages)
    sorted_percentage_df = percentage_df.sort_values(by='Percentage_of_Total', ascending=False)

    return sorted_percentage_df
```

```python
def drop_columns(df: pd.DataFrame, columns_to_drop: list) -> pd.DataFrame:
    """Drop specified columns from a DataFrame.

    Args:
    - df (pd.DataFrame): The DataFrame from which to drop columns.
    - columns_to_drop (list): A list of column names to be dropped.

    Returns:
    - pd.DataFrame: A new DataFrame with specified columns removed.
    """
    # Drop the specified columns
    df_dropped = df.drop(columns=columns_to_drop, errors='ignore')
    return df_dropped
```

```python
# helper function for plotting graphs
def plot_distributions(data: pd.DataFrame,columns_to_plot=None) -> None:
    """
    Plot distributions for each column in the provided DataFrame.

    This function iterates over the specified columns (or each column if none are specified)
    in the DataFrame and plots the distribution.For numerical columns, a histogram with a 
    Kernel Density Estimate (KDE) is plotted. For categorical columns, a countplot is used.

    Parameters:
    - data (pandas.DataFrame): The DataFrame containing the data for which distributions
                               are to be plotted.
    - columns_to_plot (list, optional): A list of column names to plot. If None, all
                                        columns will be plotted.

    Returns:
    - None: This function does not return any value. It directly displays the plots.
    """
    if columns_to_plot is None:
        columns_to_plot = data.columns

    for column in columns_to_plot:
        if column in data.columns:
            plt.figure(figsize=(10,4))

        # Check if the column is numeric or categorical
            if data[column].dtype == 'object' or data[column].dtype == 'bool':
                # For categorical data, use countplot
                sns.countplot(y=column, data=data)
                plt.title(f'Distribution of {column}')
            else:
                # For numerical data, use histogram
                sns.histplot(data[column], kde=True)
                plt.title(f'Distribution of {column}')
            
            plt.show() 
        else:
            print(f"Column '{column}' not found in DataFrame.")
```

#### Outlier Handling
* For which columns should we apply outlier detect: As you can see above graphs, there is no outlier for categorical variables such as 'is_a_buyer', 'pro_user' etc. So we will handle only the outliers of 'mrp','price','product_rating_count' variables.
* Which method should we use:
  * IQR is based on the median and quadrants of the data, so it is useful in data sets that are not normally distributed or symmetric
  * Z-Score Method uses the median when identifying outliers, so it may be appropriate for normal undistributed datasets
  * As you can see from the graphs above, our data has not a normal distribution (especially 'mrp','price','product_rating_count' columns), so that we should use IQR method for outlier detection.

```python
# helper function for outlier detecting 
features = ['mrp','price','product_rating_count']
def detect_outliers(series):
    # Quartile of features
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    # Interquartile range calculation
    IQR = Q3 - Q1
    mild_outlier_threshold = 1.5 * IQR
    # Selecting outliers
    outliers = ((series < (Q1 - mild_outlier_threshold)) | (series > (Q3 + mild_outlier_threshold))).sum()
    
    return outliers

# Store the counts of outliers in a new DataFrame
outlier_counts = pd.DataFrame(columns=['Outliers'])


for column in features:
    # Traverse within numberical features to find exact count of outliers for each feature.
    outliers= detect_outliers(df_dropped[column])
    # Adding them to dataframe
    outlier_counts.loc[column] = [outliers]

outlier_counts 
```

```python
# outlier handling by median imputation
for feature in features:
    # Quartile of features
    q1 = df_dropped[feature].quantile(0.25)
    q3 = df_dropped[feature].quantile(0.75)
    # IQR calculation
    iqr = q3-q1
    # Upper and lower bounds for values in data set that is acceptable
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    df_dropped[feature]=df_dropped[feature].mask((df_dropped[feature] < lower_fence) | (df_dropped[feature] > upper_fence), df_dropped[feature].median())
    
```

```python
# Necessary imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from typing import Tuple, List
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score,precision_recall_curve,confusion_matrix

import numpy as np
```

```python
X = data_scaled_df.drop(columns=['is_a_buyer'])
Y = data_scaled_df['is_a_buyer']
```

```python
def knn_classification(X: pd.DataFrame, Y: pd.Series, test_size: float = 0.2, p: int = 2,
                        k_range: List[int] = range(1, 31, 2)) -> Tuple[int, List[float]]:
    """
    Perform K-nearest neighbors classification with various values of K.

    Parameters:
    - X: DataFrame containing features
    - Y: Series containing the target variable
    - test_size: Proportion of the dataset to include in the test split (default: 0.2)
    - p: Minkowski distance function exponentiation (default: 2)
    - k_range: List of K values to try (default: [1, 3, 5, ..., 29])

    Returns:
    - accuracy: Accuracy score for the optimal K value
    - precision: Precision score for the optimal K value
    - recall: Recall score for the optimal K value
    - f1: F1 score for the optimal K value
    """

    # Split the dataset into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=42)

    # Error values container
    error_rate = []

    for K in k_range:
        # K-neighbors classifier with Minkowski distance
        model = KNeighborsClassifier(n_neighbors=K, p=p)
        model.fit(X_train, Y_train)
        y_pred = model.predict(X_test)
        error = np.mean(y_pred != Y_test)
        error_rate.append(error)

    # Plot the Error Graph
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, error_rate, marker='o')
    plt.title('Elbow Method For Optimal K')
    plt.xlabel('Number of Neighbors (K)')
    plt.ylabel('Error Rate')
    plt.xticks(k_range)
    plt.grid()
    plt.show()

    # Select the optimal K value
    k_optimal = k_range[error_rate.index(min(error_rate))]

    # Train the model with the optimal K value
    optimal_model = KNeighborsClassifier(n_neighbors=k_optimal, p=p)
    optimal_model.fit(X_train, Y_train)
    y_pred_optimal = optimal_model.predict(X_test)

    # Calculate and print accuracy, precision, recall, and F1 score
    accuracy = accuracy_score(Y_test, y_pred_optimal)
    precision = precision_score(Y_test, y_pred_optimal)
    recall = recall_score(Y_test, y_pred_optimal)
    f1 = f1_score(Y_test, y_pred_optimal)

    print(f"Optimal K       : {k_optimal}")
    print("analyse the performance:")
    print(f"Accuracy Score  : {accuracy}")
    print(f"Precision Score : {precision}")
    print(f"Recall Score    : {recall}")
    print(f"F1 Score        : {f1}")

    conf_matrix = confusion_matrix(Y_test, y_pred)
    print("\nConfusion Matrix:")
    
    # Plot the confusion matrix using Seaborn
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', linewidths=.5, cbar=False)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()

    return accuracy, precision, recall, f1
    
    
```

```python
# for p = 2, the distance that is measured will be euclidean distance
# our function finds the optimal K value by trying different K values (k_range)
# the test_size parameter is used to divide the dataset into training and test sets
# p parameter determines whether the Minkowski distance is Euclidean (p=2) or Manhattan (p=1)
knn_euclidean_acc,knn_euclidean_pre,knn_euclidean_rec,knn_euclidean_f1 = knn_classification(X, Y,p=2)
```

#### Logistic Regression Classifier

```python
def logistic_reg_classifier(X: pd.DataFrame, Y: pd.Series, test_size: float = 0.2):
    """
    Perform Logistic Regression classification.

    Parameters:
    - X: DataFrame containing features
    - Y: Series containing the target variable
    - test_size: Proportion of the dataset to include in the test split (default: 0.2, float)

    Returns:
    - accuracy: Accuracy score of the classifier
    - precision: Precision score of the classifier
    - recall: Recall score of the classifier
    - f1: F1 score of the classifier
    
    This function trains a Logistic Regression classifier, evaluates its performance, and displays the following:
    - Error rate
    - Accuracy score
    - Confusion matrix
    - Precision score
    - Recall score
    - F1 score
    - Precision-Recall curve
    """
    # Split the dataset into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=42)
    
    # Create a Logistic Regression classifier
    logreg = LogisticRegression()

    # Train the classifier
    logreg.fit(X_train, Y_train)

    # Predict on test data
    y_pred_logistic = logreg.predict(X_test)
    
    # Calculate error rate
    error = np.mean(y_pred_logistic != Y_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(Y_test, y_pred_logistic)

    # Calculate precision, recall, and F1 scores
    precision = precision_score(Y_test, y_pred_logistic)
    recall = recall_score(Y_test, y_pred_logistic)
    f1 = f1_score(Y_test, y_pred_logistic)

    # Print error rate, accuracy, precision, recall, and F1 score
    print("analyse the performance:")
    print(f"Error rate     : {error}")
    print(f"Accuracy Score : {accuracy}")
    print(f"Precision      : {precision}")
    print(f"Recall         : {recall}")
    print(f"F1 Score       : {f1}")

    # Calculate and display confusion matrix
    conf_matrix = confusion_matrix(Y_test, y_pred_logistic)
    print("\nConfusion Matrix:")
    
    # Plot the confusion matrix using Seaborn
    plt.figure(figsize=(10, 7))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', linewidths=.5, cbar=False)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    
    # Calculate precision and recall values for the precision-recall curve
    precision, recall, _ = precision_recall_curve(Y_test, logreg.predict_proba(X_test)[:, 1])

    # Plot the precision-recall curve
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, marker='o')
    plt.title('Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.grid()
    plt.show()
    
    return accuracy, precision, recall, f1
```

## RESULT
Which evaluation metric significative for our data set (we should answer first this question before any comparing):
* Accuracy Score:  If the classes are unbalanced, the accuracy score may be insufficient. That is, if a class has many more observations, the model may be more inclined to predict this class correctly. For out data set the classes are unbalanced, we explained why in the Data Analysis part.
* Precision Score: If the cost of false positives (FP) is high, precision can be an important metric. But for our data set, cost of false positives is not  that high.
* Recall Score: If the cost of false negatives (FN) is high, recall can be an important metric. For our data, the cost of FN is very slightly higher than FP. Because, it costs more to assume that someone who is a customer, is not a customer than to assume that someone who is not a customer, is a customer. Because the survival of the product is ensured by the customers who buy it, since the ideas and comments of the real customer will be important for the product, the incorrect prediction of the real customer will cause serious losses for the product.
* F1 Score: It is the harmonic average of precision and recall. It provides a good performance measurement in unstable data sets. F1 score is suitable for our data set too. Because even though we decide, between recall and precision, that recall is more important for our data , the difference in importance between the two is very small. Therefore, the F1 score, which is the harmonic average of the two, is the metric that will make the final decision.
